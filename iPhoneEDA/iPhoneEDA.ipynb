## iPhone Exploratory Data Analysis

```python
## importing libaries for our EDA 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## reading our CSV file that contains the data
df = pd.read_csv("Documents/apple_products.csv")
```

```python
## only checking first 5 rows
df.head()
```

```python
## counting how many rows we have in our data
df.count()
```

```python
## finding the highest price of iPhones
df['Mrp'].max()
```

```python
## finding the lowest price of all iPhones
df['Mrp'].min()
```

```python
## the specific iPhone that has the highest price
df[df['Mrp'] == 149900]
```

```python
## the specific iPhone that has the lowest price
df[df['Mrp'] == 39900]
```

```python
## looking to see which iPhones have a price greater than what's listed 
df[df['Mrp'] >= 100000]
```

```python
## selecting which data we want to look at closely
selected_data = df.loc[:, ['Product Name', 'Brand', 'Mrp', 'Star Rating']]

## looking at first couple rows
selected_data.head()
```

```python
## looking at which ratings have exactly 4.5
selected_data[selected_data['Star Rating'] <= 4.5]

## looking at the lowest rating 
min = selected_data['Star Rating'].min()
print(min)
```

```python
## creating a histogram of price of each iPhone
selected_data.head()

## getting the substring of the iPhone name (ex: iPhone 8)
name = selected_data['Product Name'].str[6:15]

## bar plot we are creating for the ratings of iPhones
plt.bar(name, selected_data['Star Rating'], color = 'navy', edgecolor = 'black')
plt.xlabel('iPhone')
plt.ylabel('Rating')
plt.title('Ratings Given for Each iPhone')
plt.show()
```

```python
## storing the head of selected data into new variable
first5 = selected_data.head()

## creating our bar chart

plt.bar(selected_data['Product Name'].str[6:15], selected_data['Mrp'], color = 'blue', edgecolor = 'black')
plt.xlabel('iPhone')
plt.ylabel('Price')
plt.title('Price of iPhones (Bar Chart)')
plt.xticks(fontsize = 10)

plt.show()
```

```python
## first 5 rows we will be using
first5whole = df.head()
first5whole.head()

## generating values for heatmap
values = df[['Mrp', 'Number Of Ratings', 'Number Of Reviews']].corr()
sns.heatmap(values, annot = True)
plt.title('Heatmap of Price, # of Ratings, and # of Reviews')

##showing the heatmap
plt.show()
```

```python
df.head()

pairData = df[['Product Name', 'Mrp', 'Number Of Ratings', 'Number Of Reviews',]]
pairplot = sns.pairplot(pairData)

pairplot.map_upper(sns.scatterplot, color = 'blue')
pairplot.map_lower(sns.scatterplot, color = 'red')

plt.title('Pair Plot of Data')
plt.show()
```
